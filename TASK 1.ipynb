```python
def caesar_cipher(text, shift, mode, preserve_non_alpha):
    result = ""
    for char in text:
        if char.isalpha():
            # Determine whether to encrypt or decrypt based on the mode
            if mode == 'encrypt':
                shifted = ord(char) + shift
            elif mode == 'decrypt':
                shifted = ord(char) - shift
            # Adjust for uppercase letters
            if char.isupper():
                if shifted > ord('Z'):
                    shifted -= 26
                elif shifted < ord('A'):
                    shifted += 26
            # Adjust for lowercase letters
            elif char.islower():
                if shifted > ord('z'):
                    shifted -= 26
                elif shifted < ord('a'):
                    shifted += 26
            result += chr(shifted)
        elif not char.isspace() and preserve_non_alpha:
            result += char
    return result

def brute_force_decrypt(ciphertext):
    print("Attempting brute-force decryption:")
    for shift in range(1, 26):
        decrypted_text = caesar_cipher(ciphertext, shift, 'decrypt', False)
        print(f"Shift {shift}: {decrypted_text}")

def main():
    while True:
        mode = input("Enter 'encrypt' or 'decrypt': ").lower()
        if mode not in ['encrypt', 'decrypt']:
            print("Invalid mode! Please enter 'encrypt' or 'decrypt'.")
            continue
        text = input("Enter your message: ")
        shift = int(input("Enter the shift value (positive for right shift, negative for left shift): "))
        preserve_non_alpha = input("Preserve non-alphabetic characters? (yes/no): ").lower() == 'yes'

        try:
            result = caesar_cipher(text, abs(shift) % 26, mode, preserve_non_alpha)
            print("Result:", result)
        except ValueError:
            print("Invalid shift value! Shift value must be an integer.")
            continue

        if mode == 'decrypt':
            choice = input("Do you want to attempt brute-force decryption? (yes/no): ")
            if choice.lower() == 'yes':
                brute_force_decrypt(text)

        choice = input("Do you want to continue? (yes/no): ")
        if choice.lower() != 'yes':
            break

if __name__ == "__main__":
    main()
```
